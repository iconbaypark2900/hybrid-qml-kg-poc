# Jupyter AI — chat interface (quick reference)

Official user guide: [Jupyter AI — The chat interface](https://jupyter-ai.readthedocs.io/en/latest/users/index.html#the-chat-interface).

Jupyter AI sends prompts to **third-party** model providers. Review each provider’s **privacy policy** and **pricing** before use.

## Prerequisites

- **JupyterLab 4** (recommended), or **Jupyter Notebook 7.5+**.
- Python **3.9–3.13** (see [Users — Prerequisites](https://jupyter-ai.readthedocs.io/en/latest/users/index.html#prerequisites)).
- Install the package, then **restart** the Jupyter server so the lab extension loads.

```bash
pip install "jupyter-ai[all]"==<version>   # e.g. latest from docs; quote [all] for zsh
```

Provider-specific extras (e.g. `boto3` for Bedrock) are listed in the [Model providers](https://jupyter-ai.readthedocs.io/en/latest/users/index.html#model-providers) section.

## Open the chat UI

**JupyterLab:** After starting `jupyter lab`, click the **chat** icon in the **left** side panel. You can right-click the panel icon to move the chat to the other side.

**Jupyter Notebook 7.5+:** **View → Left sidebar → Show Jupyter AI Chat**.

## Choose a model

1. Open **Settings → Jupyternaut settings**.
2. Pick a **language model** (search works). Jupyter AI uses **LiteLLM**; many providers need an **API key** set in the same settings panel.
3. Optional: set **model parameters** (v3+).

## Use the chat

- **+ Chat** — start a named conversation; chats persist as `<chat_name>.chat` server documents (context for the LLM).
- **@-mention** `@Jupyternaut` (default persona) in the prompt area.
- **ENTER** sends; **SHIFT+ENTER** adds a newline (defaults; configurable in chat settings).
- **Include selection** — attach highlighted notebook content (may increase tokens/cost).
- **`/generate`** — ask for a new notebook from a prompt (can take minutes; keep JupyterLab open).
- Drag a **cell with an error** into the chat to ask for a fix (see docs for `/fix`-style flows).

Memory: the backend typically uses the **last two exchanges** as context (see [Additional details](https://jupyter-ai.readthedocs.io/en/latest/users/index.html#additional-details-about-the-chat-interface)).
